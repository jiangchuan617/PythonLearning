## MatplotLib

### matplotlib 的简单绘图-plot

```python
import matplotlib.pyplot as plt
import numpy as np
# 在jupyter中需要添加下面的语句
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
a = [2, 3, 4]
b = [4, 5, 6]
plt.plot(a, b)
# plt.show()
```

```python
t = np.arange(0,2,0.1)
s = np.sin(t*np.pi)
plt.plot(t,s,'r--',label = "aaa")
plt.plot(t*2,s,'--',label = "bbb")
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.title("this is a demo")
plt.legend( )
```

### matplotlib 的简单绘图-subplot

```python
x = np.linspace(0, 5)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2);
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,'b--',label = 'sin(pi*x)')
plt.ylabel("y1 value")
plt.subplot(2,1,2)
plt.plot(x,y2,'r--',label = 'sin(pi*2x)')
plt.ylabel("y2 value")
plt.xlabel("x value")
# plt.title("this is x-y value")
# plt.legend()
```

```python
fig, ax = plt.subplots(2,2)
```

```python
ax
```

```python
fig, ax = plt.subplots(2,2)
ax[0][0].plot(x,y1)
ax[0][1].plot(x,y2)
plt.show()
```

### pandas 绘图之Series

```python
from pandas import Series
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(kind = "line",grid = True,label = "S1",title ='this is Series',style = '--')
plt.legend()
plt.show()
```

```python
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(kind = "line",grid = True,label = "S1",title ='this is Series',style = '--')
s2.plot(label = "S2")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2,1)
s1.plot(ax = ax[0],label = "S1")
s2.plot(ax = ax[1],label = "S2")
plt.legend()
```

### pandas 绘图之DataFrame

```python
from pandas import DataFrame
```

```python
df = DataFrame(np.random.randint(1, 10, 40).reshape(10, 4),
               columns=['A', 'B', 'C', 'D'])
```

```python
df.plot()
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'barh')
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df.plot(kind='area')
```

```python
df.iloc[5].plot()
```

```python
for i in df.index:
    df.iloc[i].plot(label = str(i))
plt.legend(loc ='upper right')
```

```python
df.T.plot()
plt.legend(loc ='upper right')
```
