### seaborn 设置图形显示效果

```python
x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25;
```

```python
def sinplot():
    plt.plot(x,y1)
    plt.plot(x,y2)
```

```python
sinplot()
```

```python
styles = ['darkgrid','dark','white','whitegrid','ticks']
sns.set_style(styles[0])
sinplot()
```

```python
sns.axes_style()
```

```python
sns.set_style(styles[0], {'grid.color': 'red'})
sinplot()
```

```python
sns.set() # 清空更改风格
sinplot()
```

```python
contexts = ['paper', 'notebook', 'talk', 'poster']
```

```python
sns.set_context(contexts[2])
sinplot()
```

```python
sns.plotting_context()
```

```python
sns.set_context(contexts[2], rc={'grid.linewidth': 4.0})
sinplot()
```

```python
sns.set() # 恢复默认
```

### seaborn 调色功能

```python
def sinplot():
    x = np.linspace(0, 14, 100)
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x, np.sin(x+i)*(i+0.75), label='sin(x+%s)*(%s+0.75)' % (i, i))
    plt.legend()
```

```python
sinplot()
```

```python
# 调色板
sns.color_palette()  # RGB
```

```python
sns.palplot(sns.color_palette())
```

```python
pal_style = ["deep",'muted','pastel','bright','dark','colorblind']
# sns.palplot(sns.color_palette("deep"))
sns.palplot(sns.color_palette(pal_style[5]))
```
