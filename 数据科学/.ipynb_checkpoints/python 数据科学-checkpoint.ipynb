### 数据分箱技术Binning

```python
score_list = np.random.randint(25, 100, 20)
```

```python
bins = [0,59,70,80,100];
```

```python
score_cut = pd.cut(score_list,bins)
pd.value_counts(score_cut)
```

```python
df = DataFrame()
df['score'] = score_list
df['student'] = [pd.util.testing.rands(3) for i in range(20)]
df
```

```python
df['categories'] = pd.cut(df['score'],bins)
df
```

```python
df['categories'] = pd.cut(df['score'],bins,labels=['Low','Ok','Good','Great'])
df
```

### 数据分组技术GroupBy

```python
df = pd.DataFrame({"date": pd.date_range("2021-03-01", periods=30),
                   "city": ["BJ"]*12+["SH"]*10+['GZ']*8,
                   "temperature": np.random.randint(5, 38, 30),
                   "wind": np.random.randint(1, 7, 30)}
                  )
```

```python
g = df.groupby('city')
```

```python
g.groups
```

```python
df_bj = g.get_group("BJ")
```

```python
df_bj.mean()
```

```python
g.mean()
```

```python
list(g)
```

```python
dict(list(g))['BJ']
```

```python
for name,group_df in g:
    print(name)
    print(group_df)
```

```python
g_new = df.groupby(['city', 'wind'])
```

```python
g_new.groups
```

```python
g_new.get_group(("BJ",2))
```

```python
for (name_1,name_2),group_df in g_new:
    print(name_1,name_2)
    print(group_df)
```

### 数据聚合技术 Aggregation

```python
df = pd.DataFrame({"date": pd.date_range("2021-03-01", periods=30),
                   "city": ["BJ"]*12+["SH"]*10+['GZ']*8,
                   "temperature": np.random.randint(5, 38, 30),
                   "wind": np.random.randint(1, 7, 30)}
                  )
g = df.groupby('city');
```

```python
g.describe()
```

```python
g.agg(max)
```

```python
g.agg("max")
```

```python
def foo(attr):
    return attr.max()-attr.min()
    
```

```python
g.agg(foo)
```

### 透视表

```python
df = pd.read_excel("data/sales-funnel.xlsx")
```

```python
pd.pivot_table(df, index=["Name"])
```

```python
pd.pivot_table(df, index=["Name"],aggfunc="sum")
```

```python
pd.pivot_table(df, index=["Name","Manager","Rep"])
```

```python
pd.pivot_table(df, index=["Manager","Rep"])
```

```python
pd.pivot_table(df, index=["Manager", "Rep"], values=['Price'])
```

```python
pd.pivot_table(df, index=["Manager", "Rep"], values=['Price'], aggfunc="sum")
```

```python
pd.pivot_table(df, index=["Manager", "Rep"], values=['Price', 'Quantity'], aggfunc="sum")
```

```python
pd.pivot_table(df, index=["Manager", "Rep"], values=[
               'Price', 'Quantity'], columns=['Product'], aggfunc="sum")
```

```python
pd.pivot_table(df, index=["Manager", "Rep"], values=[
               'Price', 'Quantity'], columns=['Product'], fill_value=0, aggfunc="sum")
```

### matplotlib 的简单绘图-plot

```python
import matplotlib.pyplot as plt
import numpy as np
# 在jupyter中需要添加下面的语句
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
a = [2, 3, 4]
b = [4, 5, 6]
plt.plot(a, b)
# plt.show()
```

```python
t = np.arange(0,2,0.1)
s = np.sin(t*np.pi)
plt.plot(t,s,'r--',label = "aaa")
plt.plot(t*2,s,'--',label = "bbb")
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.title("this is a demo")
plt.legend( )
```

### matplotlib 的简单绘图-subplot

```python
x = np.linspace(0, 5)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2);
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,'b--',label = 'sin(pi*x)')
plt.ylabel("y1 value")
plt.subplot(2,1,2)
plt.plot(x,y2,'r--',label = 'sin(pi*2x)')
plt.ylabel("y2 value")
plt.xlabel("x value")
# plt.title("this is x-y value")
# plt.legend()
```

```python
fig, ax = plt.subplots(2,2)
```

```python
ax
```

```python
fig, ax = plt.subplots(2,2)
ax[0][0].plot(x,y1)
ax[0][1].plot(x,y2)
plt.show()
```

### pandas 绘图之Series

```python
from pandas import Series
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(kind = "line",grid = True,label = "S1",title ='this is Series',style = '--')
plt.legend()
plt.show()
```

```python
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
s1.plot(kind = "line",grid = True,label = "S1",title ='this is Series',style = '--')
s2.plot(label = "S2")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2,1)
s1.plot(ax = ax[0],label = "S1")
s2.plot(ax = ax[1],label = "S2")
plt.legend()
```
